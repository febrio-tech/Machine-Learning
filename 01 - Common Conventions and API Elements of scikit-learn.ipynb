````python

import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
````


## Ringkasan Chapter

Scikit-learn dirancang berdasarkan prinsip **consistency**, **simplicity**, **modularity**, dan **reusability**. Hampir seluruh algoritma machine learning di dalam library ini menggunakan pola API yang sama sehingga lebih mudah dipelajari dan digunakan.

## Filosofi Desain scikit-learn

Scikit-learn memiliki desain yang konsisten di seluruh algoritma. Baik menggunakan Linear Regression, Random Forest, KNN, maupun model lain, pola penggunaannya hampir sama.

| Prinsip | Penjelasan |

## Memahami Estimator

Estimator adalah objek yang belajar dari data. Hampir semua algoritma machine learning dalam scikit-learn merupakan estimator.

Estimator biasanya memiliki dua metode utama:

- `fit()` → belajar dari data
- `predict()` → menghasilkan prediksi


````python

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([1,2,3,3.5,5])

model = LinearRegression()
model.fit(X,y)

prediksi = model.predict([[6],[7]])

print(prediksi)
````


### Alur Kerja Estimator

| Tahap | Fungsi |
|---------|---------|
| Membuat model | `LinearRegression()` |
| Training | `fit(X,y)` |
| Prediksi | `predict(X_baru)` |
| Evaluasi | `score(X,y)` |



## Fit, Predict, dan Fit Predict

Pada supervised learning, data training digunakan untuk proses `fit()`, sedangkan data baru digunakan pada `predict()`.

Pada beberapa algoritma unsupervised seperti K-Means tersedia shortcut berupa `fit_predict()`, yaitu melakukan training dan menghasilkan label cluster sekaligus.


````python

from sklearn.cluster import KMeans

X = np.array([[1],[2],[3],[4],[5]])

kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

print(labels)
````


## Transformer dan Method Transform

Transformer digunakan untuk mengubah data sebelum masuk ke model.

Contoh transformasi:

- Scaling
- Normalisasi
- Encoding
- Feature Engineering

Transformer memiliki dua metode utama:

- `fit()` → mempelajari parameter transformasi
- `transform()` → menerapkan transformasi


````python

X = np.array([[1,2],
              [3,4],
              [5,6]])

scaler = StandardScaler()

scaler.fit(X)

X_scaled = scaler.transform(X)

print(X_scaled)
````


### Perbedaan fit() dan transform()

| Method | Fungsi |
|----------|---------|
| fit() | Menghitung parameter dari data |
| transform() | Mengubah data menggunakan parameter tadi |
| fit_transform() | Menggabungkan fit dan transform |


````python

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled)
````


## Mengapa Transformasi Penting?

Banyak algoritma machine learning sensitif terhadap skala data.

Contoh algoritma yang sangat terbantu dengan scaling:

- KNN
- Logistic Regression
- SVM
- Neural Network

Transformasi juga membantu mengurangi pengaruh perbedaan satuan antar fitur.



## Custom Estimator dan Custom Transformer

Salah satu keunggulan scikit-learn adalah kemudahan membuat estimator sendiri.

Custom estimator harus mengikuti API yang sama sehingga tetap kompatibel dengan:

- Pipeline
- GridSearchCV
- Cross Validation


````python

from sklearn.base import BaseEstimator, TransformerMixin

class AddOneTransformer(BaseEstimator, TransformerMixin):

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X + 1

X = np.array([[1],[2],[3]])

custom = AddOneTransformer()

print(custom.fit_transform(X))
````


## Pipeline dan Workflow Automation

Pipeline memungkinkan beberapa langkah preprocessing dan modeling digabung menjadi satu workflow.

Keuntungan pipeline:

- Mengurangi kesalahan preprocessing
- Meningkatkan reproducibility
- Memudahkan deployment
- Mendukung cross-validation dan grid search



### Struktur Pipeline

```text
Data
 ↓
Transformer
 ↓
Transformer
 ↓
Model
 ↓
Prediksi
```


````python

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

data = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    data.data,
    data.target,
    random_state=42
)

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=5000))
])

pipe.fit(X_train, y_train)

print("Accuracy:", pipe.score(X_test, y_test))
````

## Common Attributes dan Methods

Sebagian besar model scikit-learn memiliki atribut dan method yang seragam.

| Atribut / Method | Fungsi |
|------------------|---------|
| coef_ | Koefisien model |
| intercept_ | Nilai intercept |
| predict() | Prediksi |
| score() | Evaluasi model |
| get_params() | Melihat parameter |
| set_params() | Mengubah parameter |


````python

model = LinearRegression()

model.fit(X, np.array([1,2,3]))

print("Coef :", model.coef_)
print("Intercept :", model.intercept_)
print("Score :", model.score(X, np.array([1,2,3])))
````


## Hyperparameter Tuning

Hyperparameter adalah parameter yang ditentukan sebelum model dilatih.

Contoh:

| Model | Hyperparameter |
|---------|---------------|
| Random Forest | n_estimators |
| KNN | n_neighbors |
| SVM | C, gamma |
| Decision Tree | max_depth |


````python

model = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=42
)

print(model.get_params()["n_estimators"])
print(model.get_params()["max_depth"])
````


## Grid Search

Grid Search digunakan untuk mencari kombinasi hyperparameter terbaik secara otomatis.


````python

from sklearn.datasets import load_iris

iris = load_iris()

param_grid = {
    "n_estimators":[50,100],
    "max_depth":[3,5,None]
}

grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=3
)

grid.fit(iris.data, iris.target)

print("Best Params:", grid.best_params_)
print("Best Score :", grid.best_score_)
````
