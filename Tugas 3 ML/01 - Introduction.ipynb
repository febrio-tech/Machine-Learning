## Mathematical Proofs Versus Intuition from Coding

Soft proof bukan pengganti proof formal. Namun, soft proof dapat membantu pembaca memahami pola matematika tanpa langsung tenggelam dalam notasi dan argumentasi abstrak.

## Contoh Soft Proof: Sifat Komutatif Perkalian

Chapter ini memberikan contoh sederhana tentang sifat komutatif perkalian:

$$
a \times b = b \times a
$$

Ide eksperimennya adalah membuat dua angka acak, lalu memeriksa apakah selisih antara $a \times b$ dan $b \times a$ bernilai nol. Jika hasilnya nol pada berbagai percobaan, pembaca dapat membangun intuisi bahwa urutan perkalian tidak mengubah hasil.

````python
import numpy as np

a = np.random.randn()
b = np.random.randn()

a * b - b * a
````

## Penjelasan Output Kode

Kode di atas menghasilkan dua angka acak, yaitu `a` dan `b`. Kemudian kode menghitung:

```python
a*b - b*a
```

Jika sifat komutatif berlaku, maka hasilnya adalah `0.0` atau sangat dekat dengan nol karena keterbatasan presisi komputer. Contoh ini menunjukkan bagaimana kode dapat digunakan untuk menguji dan memahami konsep matematika secara numerik.

Namun, hasil tersebut tidak boleh dianggap sebagai bukti formal untuk semua kemungkinan angka. Fungsinya adalah sebagai alat bantu intuisi.
