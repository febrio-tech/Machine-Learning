## Vector Sets

Vector set adalah kumpulan vector yang ditulis menggunakan huruf kapital, misalnya:

$$
V = \{v_1,v_2,\ldots,v_n\}
$$

Vector set dapat berisi jumlah vector yang terbatas maupun tak terbatas. Dalam data science, dataset sering dipandang sebagai sebuah vector set karena setiap observasi dapat direpresentasikan sebagai sebuah vector.



## Linear Weighted Combination

Linear weighted combination merupakan proses mengalikan setiap vector dengan scalar tertentu kemudian menjumlahkannya.

### Rumus Linear Weighted Combination

$$
w = \lambda_1v_1 + \lambda_2v_2 + \cdots + \lambda_nv_n
$$

Keterangan:

- $w$ = vector hasil kombinasi
- $v_i$ = vector ke-$i$
- $\lambda_i$ = scalar atau weight
- Semua vector harus memiliki dimensionalitas yang sama

Konsep ini menjadi dasar bagi regresi linear, PCA, neural network, dan berbagai metode machine learning lainnya.


```python

import numpy as np

l1,l2,l3 = 1,2,-3

v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])

w = l1*v1 + l2*v2 + l3*v3
print(w)
```


## Linear Independence dan Linear Dependence

Sebuah vector set disebut **linearly dependent** jika setidaknya ada satu vector yang dapat ditulis sebagai kombinasi linear dari vector lainnya.

Sebaliknya, sebuah vector set disebut **linearly independent** jika tidak ada vector yang dapat dibentuk dari kombinasi vector lain dalam set tersebut.

Contoh:

$$
V =
\left\{
\begin{bmatrix}1\\3\end{bmatrix},
\begin{bmatrix}2\\7\end{bmatrix}
\right\}
$$

Set di atas independen karena tidak ada scalar yang dapat mengubah satu vector menjadi vector lainnya.



### Definisi Matematis Linear Dependence

Suatu vector set dikatakan dependent apabila terdapat scalar yang memenuhi:

$$
0 = \lambda_1v_1 + \lambda_2v_2 + \cdots + \lambda_nv_n
$$

dengan syarat tidak semua $\lambda_i$ bernilai nol.

Keterangan:

- $0$ = zero vector
- $\lambda_i$ = scalar
- Minimal ada satu scalar yang tidak bernilai nol

Jika persamaan tersebut memiliki solusi non-trivial, maka set bersifat dependent.



## Independence dan Zero Vector

Chapter ini menekankan bahwa setiap vector set yang mengandung zero vector otomatis menjadi dependent.

Hal ini terjadi karena:

$$
\lambda_0\mathbf{0}=0
$$

untuk sembarang nilai scalar $\lambda_0$. Oleh karena itu selalu ada solusi non-trivial yang memenuhi definisi linear dependence.



## Subspace dan Span

Subspace terbentuk dari seluruh kemungkinan linear weighted combination yang dapat dibuat dari suatu vector set.

Sedangkan span adalah proses atau mekanisme pembentukan subspace tersebut.

Secara sederhana:

- **Span** = proses membentuk kombinasi linear.
- **Subspace** = hasil dari seluruh kombinasi linear tersebut.



### Contoh Span Satu Vector

Misalkan:

$$
V =
\left\{
\begin{bmatrix}
1\\3
\end{bmatrix}
\right\}
$$

Span dari vector tersebut adalah seluruh kelipatan scalarnya:

$$
\lambda
\begin{bmatrix}
1\\3
\end{bmatrix}
$$

Hasilnya berupa garis lurus tak hingga yang melalui titik asal.


```python

import numpy as np
import matplotlib.pyplot as plt

v = np.array([1,3])

scalars = np.linspace(-4,4,100)
points = np.outer(scalars,v)

plt.scatter(points[:,0],points[:,1],s=10)
plt.grid(True)
plt.title("Span dari Satu Vector")
plt.show()
```


### Hubungan Span dan Linear Independence

Jika sebuah vector set independen, maka jumlah vector biasanya sama dengan dimensi subspace yang dibentuk.

Namun jika terdapat vector yang redundant (dependent), dimensi subspace menjadi lebih kecil daripada jumlah vector dalam set tersebut.

Contoh:

$$
\left\{
\begin{bmatrix}
1\\1\\1
\end{bmatrix},
\begin{bmatrix}
2\\2\\2
\end{bmatrix}
\right\}
$$

Meskipun terdapat dua vector, keduanya hanya membentuk subspace berdimensi satu karena vector kedua hanyalah dua kali vector pertama.



## Basis

Chapter ini menjelaskan basis sebagai seperangkat "penggaris" yang digunakan untuk mengukur atau mendeskripsikan suatu ruang.

Basis paling umum adalah basis Cartesian:

### Basis Standar pada $\mathbb{R}^2$

$$
S_2 =
\left\{
\begin{bmatrix}1\\0\end{bmatrix},
\begin{bmatrix}0\\1\end{bmatrix}
\right\}
$$

### Basis Standar pada $\mathbb{R}^3$

$$
S_3 =
\left\{
\begin{bmatrix}1\\0\\0\end{bmatrix},
\begin{bmatrix}0\\1\\0\end{bmatrix},
\begin{bmatrix}0\\0\\1\end{bmatrix}
\right\}
$$

Basis standar terdiri dari unit vector yang saling orthogonal.



## Mengapa Basis Penting?

Basis digunakan untuk mendeskripsikan data dalam suatu ruang. Banyak metode data science sebenarnya berusaha mencari basis baru yang lebih baik untuk merepresentasikan data.

Beberapa contoh yang disebutkan dalam chapter:

- Principal Component Analysis (PCA)
- Independent Component Analysis (ICA)
- Factor Analysis
- Singular Value Decomposition (SVD)
- Linear Discriminant Analysis (LDA)

Metode-metode tersebut pada dasarnya mencari basis yang paling sesuai untuk tujuan tertentu.



## Definisi Formal Basis

Suatu vector set dapat menjadi basis bagi suatu subspace apabila memenuhi dua syarat:

### 1. Span

Vector set harus mampu menjangkau seluruh subspace yang ingin direpresentasikan.

### 2. Linear Independence

Setiap vector harus independen sehingga setiap titik memiliki representasi yang unik.

Secara ringkas:

$$
\text{Basis} = \text{Span} + \text{Linear Independence}
$$
